## PYSEOBNR

The pyseobnr package allows the user to solve for the EOB dynamics and evaluate the waveform on the solution of the EOB dynamics, for different model of the SEOB family of waveforms.

```python
from pyseobnr.generate_waveform import generate_modes_opt
import matplotlib.pyplot as plt

q = 5.3
chi_1 = 0.9
chi_2 = 0.3
omega0 = 0.0137 # This is the orbital frequency in geometric units with M=1
t,modes = generate_modes_opt(q,chi_1,chi_2,omega0)
```

```python
plt.figure(figsize=(10, 6))
for mode in modes:
    plt.plot(t,np.real(modes[mode]))
plt.grid()
plt.show()
```

## FEW

FEW is a framework to generate waveforms for extreme mass ratio inspirals (EMRIs) from self-force models based on a multiscale expansion.

```python
import numpy as np
import matplotlib.pyplot as plt
from few.waveform import (
    FastSchwarzschildEccentricFlux
)

td_gen = FastSchwarzschildEccentricFlux()
```

```python
# define the intrinsic parameters
m1 = 1e6  # central object mass in solar masses
m2 = 1e5  # secondary object mass in solar masses
p0 = 12.0  # initial separation
e0 = 0.0  # circular
theta = 0.0
phi = 0.0

parameters = [m1, m2, p0, e0, theta, phi]

T = 0.1        # Observation time (years)
dt = 0.1        # Time sampling (seconds)
specific_modes = [(2, 2, 0, 0)]

# create TD signal
waveform_td = td_gen(*parameters,dt=dt,mode_selection=specific_modes)
time_array = np.arange(len(waveform_td)) * dt
```

```python
plt.figure(figsize=(10, 6))
plt.plot(time_array, np.real(waveform_td))
plt.show()
```

## Compare SEOB with NR

```python
# Only extract the 22 mode
import sxs
from pyseobnr.generate_waveform import generate_modes_opt
import matplotlib.pyplot as plt
import numpy as np
```

```python
sim = sxs.load("SXS:BBH:1108", auto_supersede=True)
tNR = sim.h.t
h22NR = sim.h.data[:,0]

q = sim.metadata["reference_mass_ratio"]
chi1 = sim.metadata["reference_dimensionless_spin1"][2]
chi2 = sim.metadata["reference_dimensionless_spin2"][2]
Omega = sim.metadata["reference_orbital_frequency"][2]

tSEOB,seobmodes = generate_modes_opt(q,chi1,chi2,Omega, approximant='SEOBNRv5HM')
hSEOB22 = seobmodes['2,2']
```

```python
plt.figure(figsize=(10, 6))
plt.plot(tNR, np.real(h22NR))
plt.plot(tSEOB, np.real(hSEOB22))
plt.grid()
plt.show()
```

```python
# Perform a time-shift to align the waveforms at their peak
tSEOB_shift = tSEOB[np.argmax(np.abs(hSEOB22))]
tNR_shift = tNR[np.argmax(np.abs(h22NR))]

plt.figure(figsize=(10, 6))
plt.plot(tNR - tNR_shift, np.real(h22NR))
plt.plot(tSEOB - tSEOB_shift, np.real(hSEOB22))
plt.grid()
plt.show()
```

```python
# Perform a phase-shift to align the waveforms at their peak
phiSEOB = np.angle(hSEOB22[np.argmax(np.abs(hSEOB22))])
phiNR = np.angle(h22NR[np.argmax(np.abs(h22NR))])

plt.figure(figsize=(10, 6))
plt.plot(tNR - tNR_shift, np.real(h22NR * np.exp(-1j*phiNR)))
plt.plot(tSEOB - tSEOB_shift, np.real(hSEOB22 * np.exp(-1j*phiSEOB)),linestyle='--')
plt.xlim(-3000,100)
plt.grid()
plt.show()
```
