```python
sim = sxs.load("SXS:BBH:2480", auto_supersede=True)
```

```python
# Extract the simulation metadata
q = sim.metadata["reference_mass_ratio"]
chi1 = sim.metadata["reference_dimensionless_spin1"][2]
chi2 = sim.metadata["reference_dimensionless_spin2"][2]
Omega = sim.metadata["reference_orbital_frequency"][2]
tref = sim.metadata["reference_time"]
m1 = sim.metadata["reference_mass1"]
m2 = sim.metadata["reference_mass2"]
```

```python
mask_tref = sim.h.t > tref

plt.figure(figsize=(10, 6))
plt.plot(sim.h.t[mask_tref], sim.h.data.real[mask_tref])
plt.vlines(tref, ymin=-0.1, ymax=0.1, color="k", linestyle="--", label=r"$t_{\rm ref}$")
plt.xlabel(r"$t [M]$")
plt.ylabel(r"$Re[h_{\ell m}][M]$")
plt.grid()
```

```python
FNR = np.zeros_like(sim.h.t)
xNR = np.zeros_like(sim.h.t)

for i in range(len(sim.h.LM)):
    FNRlm = np.zeros_like(sim.h.t)
    FNRlm = np.abs(np.gradient(sim.h.data[:,i], sim.h.t))**2./(16*np.pi)
    FNR += FNRlm

t_peak = sim.h.t[np.argmax(np.abs(sim.h.data[:,0]))]
mask_tref_tpeak = (sim.h.t > tref) & (sim.h.t < t_peak)

phi22NR = np.unwrap(np.angle(sim.h.data[mask_tref_tpeak,0]))/2. #Index 0 corresponds to the (2,-2) mode, /2 to go from WF to orbital phase
xNR = (np.gradient(phi22NR, sim.h.t[mask_tref_tpeak]))**(2./3.)
FNR = FNR[mask_tref_tpeak]
```

```python
plt.plot( xNR,FNR)
```

```python
plt.plot(sim.h.t[mask_tref_tpeak], xNR)
plt.xlabel(r"$t [M]$")
plt.ylabel(r"$x [/]$")
plt.grid()
plt.show()
```

```python
from PNFlux import F_4p5PN

nu = q/(1+q)**2.
FPN = F_4p5PN(xNR, nu)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(xNR, FNR)
plt.plot(xNR, FPN)
plt.xlabel(r"$x [/]$")
plt.ylabel(r"$F [/]$")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(xNR, np.abs(FPN-FNR))
plt.plot(xNR, 50*xNR**10.)
plt.xscale("log")
plt.yscale("log")
plt.xlabel(r"$x [/]$")
plt.ylabel(r"$|F_{4.5PN} - F_{NR}| [/]$")
plt.grid()
plt.show()
```

```python
from pyseobnr.generate_waveform import generate_modes_opt
```

```python
tNR = sim.h.t[mask_tref]
h22NR = sim.h.data[:,4][mask_tref]

tSEOB,seobmodes = generate_modes_opt(q,chi1,chi2,Omega, approximant='SEOBNRv5HM')
hSEOB22 = seobmodes['2,2']
```

```python
# Perform a time-shift to align the waveforms at their peak
tSEOB_shift = tSEOB[np.argmax(np.abs(hSEOB22))]
tNR_shift = tNR[np.argmax(np.abs(h22NR))]
# Perform a phase-shift to align the waveforms at their peak
phiSEOB = np.angle(hSEOB22[np.argmax(np.abs(hSEOB22))])
phiNR = np.angle(h22NR[np.argmax(np.abs(h22NR))])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(tNR - tNR_shift, np.real(h22NR * np.exp(-1j*phiNR)))
plt.plot(tSEOB - tSEOB_shift, np.real(hSEOB22 * np.exp(-1j*phiSEOB)),linestyle='--')
plt.grid()
plt.show()
```

```python
from few.utils.constants import MTSUN_SI, MRSUN_SI, Gpc
```

```python
Mtot = 61 #Total mass in solar masses
D = .44 #Distance in Gpc

tNR_sec = (tNR - tNR_shift) * Mtot * MTSUN_SI
hNR = h22NR * Mtot * MRSUN_SI / (D * Gpc)

tEOB_sec = (tSEOB - tSEOB_shift) * Mtot * MTSUN_SI
hEOB = hSEOB22 * Mtot * MRSUN_SI / (D * Gpc)

fNR22_Hz = np.gradient(np.unwrap(np.angle(hNR)), tNR_sec)/(2.*np.pi)
fEOB22_Hz = np.gradient(np.unwrap(np.angle(hEOB)), tEOB_sec)/(2.*np.pi)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(tEOB_sec, fEOB22_Hz, label='EOB f22', color='blue')
plt.plot(tNR_sec, fNR22_Hz, label='NR f22', color='orange',linestyle='--')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.ylim(-230,-20)
plt.legend()
plt.grid()
plt.show()
```

```python
from few.waveform import FastSchwarzschildEccentricFlux

few_gen = FastSchwarzschildEccentricFlux()

# parameters
M = m1 * Mtot
mu = m2 * Mtot
p0 = 1/Omega**(2./3.)
e0 = 0.0
theta = 0.0  # polar viewing angle
phi = 0.0  # azimuthal viewing angle
dt = .00001  # seconds
T = 1.0  # years
mode_selection = [(2,2,0)]

h22FEW = few_gen(M, mu, p0, e0, theta, phi, dist=D, dt=dt, T=T, mode_selection=mode_selection)/(np.sqrt(5/np.pi)/2)#Correct for inclusion or not of the spin-weighted spherical harmonics coef.

tFEW_sec = tNR_sec[0] + np.arange(len(h22FEW)) * dt
h22FEW = h22FEW * np.exp(-1j*np.angle(h22FEW[0]))
h22NR = hNR * np.exp(-1j*np.angle(hNR[0]))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(tNR_sec, np.real(h22NR))
plt.plot(tFEW_sec,np.real(h22FEW),linestyle='--')
plt.show()
```

```python
Fh22FEW = np.fft.fftshift(np.fft.fft(h22FEW))
freqsFEW = np.fft.fftshift(np.fft.fftfreq(len(h22FEW), d=tFEW_sec[1]-tFEW_sec[0]))

tNR_sec_uniform = np.linspace(tNR_sec[0], tNR_sec[-1], len(h22NR))
h22NR_uniform = np.interp(tNR_sec_uniform, tNR_sec, h22NR)
#h22NR_uniform -= np.mean(h22NR_uniform)

Fh22NR = np.fft.fftshift(np.fft.fft(h22NR_uniform))
freqsNR = np.fft.fftshift(np.fft.fftfreq(len(h22NR_uniform), d=tNR_sec_uniform[1]-tNR_sec_uniform[0]))
```

```python
plt.plot(-freqsNR, np.abs(Fh22NR))
plt.plot(-freqsFEW, np.abs(Fh22FEW))
plt.xlim(1,1e4)
#plt.vlines(0, 1e-5-20, 1e-17, colors='k', linestyles='dashed')
#plt.xlim(.001,600)
plt.yscale('log')
plt.xscale('log')
```

```python
from scipy.signal.windows import tukey
```

```python
tukey_window_FEW = tukey(len(h22FEW),alpha=.2)
tukey_window_NR = tukey(len(h22NR_uniform),alpha=.2)
```

```python
plt.plot(tFEW_sec, np.real(h22FEW))
plt.plot(tFEW_sec, np.real(tukey_window_FEW * h22FEW))
plt.show()
```

```python
plt.plot(tNR_sec_uniform, np.real(h22NR_uniform))
plt.plot(tNR_sec_uniform, np.real(tukey_window_NR * h22NR_uniform))
plt.show()
```

```python
Fh22FEW_taper = np.fft.fftshift(np.fft.fft(tukey_window_FEW*h22FEW))
Fh22NR_taper = np.fft.fftshift(np.fft.fft(tukey_window_NR*h22NR_uniform))
```

```python
plt.plot(-freqsNR, np.abs(Fh22NR))
plt.plot(-freqsFEW, np.abs(Fh22FEW))
plt.plot(-freqsFEW, np.abs(Fh22FEW_taper))
plt.plot(-freqsNR, np.abs(Fh22NR_taper))
plt.xlim(1e1,1e3)
plt.ylim(1e-21,1e-17)
plt.yscale('log')
plt.xscale('log')
```
